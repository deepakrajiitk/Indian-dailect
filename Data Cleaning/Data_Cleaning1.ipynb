# Importing libraries

```python
import pandas as pd
```

# Importing datasets

```python
df1 = pd.read_csv("DDW-0000C-13.csv")
```

```python
df1.head()
```

# Dropping irrelevant columns and rows 
# setting column names

```python
df1.columns = ['Code-Name','State-Code','District-Code','Name','Age-Group','Persons-total','Males-total','Females-total','Persons-rural','Males-rural','Females-rural','Persons-urban','Males-urban','Females-urban']
df1.drop(df1.index[:6], inplace=True)
df1.drop(columns=['Code-Name','District-Code'],inplace=True)
df1.drop(df1[df1['Age-Group'].isin(['0','1','2','3','4'])].index, inplace=True)
df1.loc[df1['Name']!='INDIA','Name'] = df1.loc[df1['Name']!='India','Name'].str[8:-5]
df1.fillna("INDIA",inplace=True)
df1.reset_index(inplace=True,drop=True)
```

```python
df1['Age-Group'].replace("All ages",'Total', inplace=True)
df1['Age-Group'].replace("100+",'100',inplace=True)
temp1 = df1[df1['Age-Group']=='Total']
temp2 = df1[df1['Age-Group']=='Age not stated']
df1.drop(df1[(df1['Age-Group']=='Total') | (df1['Age-Group']=='Age not stated')].index, inplace=True)
```

# Changing data types for error free calculation

```python
df1 = df1.apply(pd.to_numeric, errors='ignore')
```

# Combining age groups into single rows

```python
labels = ["5-9","10-14","15-19","20-24","25-29","30-49","50-69","70+"]
df1 = df1.groupby(by=['State-Code',pd.cut(df1['Age-Group'], [5,10,15,20,25,30,50,70,101],labels=labels, right=False)]).agg({'Name':'first','Age-Group':'first','Persons-total':'sum','Males-total':'sum','Females-total':'sum','Persons-rural':'sum','Males-rural':'sum','Females-rural':'sum','Persons-urban':'sum','Males-urban':'sum','Females-urban':'sum'}).drop(columns=['Age-Group']).reset_index()
df1
```

```python
df1 =  df1.append([temp1,temp2])
```

```python
df1.sort_values(['State-Code','Name'],inplace=True)
```

```python
df1
```

```python
temp1 = pd.DataFrame(columns=['State-Code','Name','TRU','Age-Group','Tot-P','Tot-F','Tot-M'])
temp2 = pd.DataFrame(columns=['State-Code','Name','TRU','Age-Group','Tot-P','Tot-F','Tot-M'])
temp3 = pd.DataFrame(columns=['State-Code','Name','TRU','Age-Group','Tot-P','Tot-F','Tot-M'])

cols = ['State-Code','Name','Age-Group','Tot-P','Tot-F','Tot-M']
col1 = ['State-Code','Name','Age-Group','Persons-total','Females-total','Males-total']
col2 = ['State-Code','Name','Age-Group','Persons-rural','Females-rural','Males-rural']
col3 = ['State-Code','Name','Age-Group','Persons-urban','Females-urban','Males-urban']

temp1[cols] = df1[col1]
temp1.fillna("Total",inplace=True)

temp2[cols] = df1[col2]
temp2.fillna("Rural",inplace=True)

temp3[cols] = df1[col3]
temp3.fillna("Urban",inplace=True)

df2 = temp1.append([temp2,temp3])
df2.sort_values(["State-Code",'TRU'],inplace=True)
```

```python
df2.head(20)
```

```python
df3 = pd.read_csv("DDW-C18-0000.csv")
```

```python
df3.head(20)
```

```python
df3.columns = ['State-Code','District-Code','Name','TRU','Age-Group','Persons-2','Males-2','Females-2','Persons-3','Males-3','Females-3']
```

```python
df3.drop(df1.index[:5], inplace=True)
df3.drop(columns=['District-Code'],inplace=True)
df3.reset_index(inplace=True,drop=True)
```

```python
df3.head()
```

# Merging datasets df2 and df3

```python
df4 = pd.merge(df2, df3, on=['Name','TRU','Age-Group']).drop(columns=['State-Code_x']).rename(columns={'State-Code_y':'State-Code'})
df4
```

```python
df4 = df4.apply(pd.to_numeric, errors='ignore')
df4.dtypes
```

```python
df4['Persons-exactly-1'] = df4['Tot-P']-df4['Persons-2']
df4['Males-exactly-1'] = df4['Tot-M']-df4['Males-2']
df4['Females-exactly-1'] = df4['Tot-F']-df4['Females-2']
df4
```

```python
df4['Persons-exactly-2'] = df4['Persons-2']-df4['Persons-3']
df4['Males-exactly-2'] = df4['Males-2']-df4['Males-3']
df4['Females-exactly-2'] = df4['Females-2']-df4['Females-3']
df4
```

```python
df4['Persons-atleast-3'] = df4['Persons-3']
df4['Males-atleast-3'] = df4['Males-3']
df4['Females-atleast-3'] = df4['Females-3']
df4
```

```python
df4.sort_values(['State-Code','TRU','Age-Group'])
```

# Final dataframe after cleaning

```python
final_df = df4[['State-Code','Name','TRU','Age-Group','Tot-P','Tot-M','Tot-F','Persons-exactly-1','Males-exactly-1','Females-exactly-1','Persons-exactly-2','Males-exactly-2','Females-exactly-2','Persons-atleast-3','Males-atleast-3','Females-atleast-3']]
final_df.sort_values(['State-Code','Name','TRU'],inplace=True)
final_df
```

```python
final_df.to_csv("cleaned_datafile.csv",index=False)
```
