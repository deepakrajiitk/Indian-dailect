# Importing libraries

```python
import pandas as pd
```

# Importing datasets

```python
df1 = pd.read_csv("DDW-C19-0000.csv")
df1.head(10)
```

# setting column names

```python
df1.columns = ['State-Code','District','Name','TRU','Education-Level','Persons-2','Males-2','Females-2','Persons-3','Males-3','Females-3']
df1
```

# Dropping irrelevant columns and rows

```python
df1.drop(df1.index[:5], inplace=True)
df1.drop(df1.index[864:], inplace=True)
df1.drop(columns=['District'],inplace=True)
df1.reset_index(inplace=True,drop=True)
```

```python
df1
```

```python
df2 = pd.read_csv("DDW-0000C-08.csv")
df2.head()
```

```python
columns = [str(i) for i in range(1,46)]
columns
df2.columns = columns
df2
```

```python
df2.drop(['1','16','17','18','43','44','45'],axis=1,inplace=True)
df2.drop(df1.index[:6], inplace=True)
df2.reset_index(inplace=True,drop=True)
```

```python
df2
```

```python
df2 = df2.apply(pd.to_numeric, errors='ignore')
df2.dtypes
```

```python
df2['28'] = df2['28'] + df2['31'] + df2['34'] + df2['37']
df2['29'] = df2['29'] + df2['32'] + df2['35'] + df2['38']
df2['30'] = df2['30'] + df2['33'] + df2['36'] + df2['39']
```

```python
df2.drop(['31','32','33','34','35','36','37','38','39'],axis=1,inplace=True)
```

```python
df2
```

```python
df2.columns = ['State-Code','District-Code','Name','TRU','Age-Group','Total-Persons','Total-Males','Total-Females','Illiterate-Persons','Illiterate-Males','Illiterate-Females','Literate-Persons','Literate-Males','Literate-Females','Below-Primary-Persons','Below-Primary-Males','Below-Primary-Females','Below-Middle-Persons','Below-Middle-Males','Below-Middle-Females','Below-Matric-Persons','Below-Matric-Males','Below-Matric-Females','Below-Graduate-Persons','Below-Graduate-Males','Below-Graduate-Females','Graduate-And-Above-Persons','Graduate-And-Above-Males','Graduate-And-Above-Females']
```

```python
df2 = df2[df2['Age-Group']=='All ages'].reset_index(drop=True)
df2.drop('Age-Group',axis = 1,inplace=True)
df2['Name'] = df2['Name'].str.replace("State - ","")
```

```python
df3 = df2.melt(id_vars=['State-Code','District-Code','Name','TRU'],value_vars=['Total-Persons','Illiterate-Persons','Literate-Persons',
'Below-Primary-Persons','Below-Middle-Persons','Below-Matric-Persons','Below-Graduate-Persons','Graduate-And-Above-Persons'],
 var_name = 'Education-Level', value_name = 'Tot-P')
```

```python
df4 = df2.melt(id_vars=['State-Code','District-Code','Name','TRU'],value_vars=['Total-Males','Illiterate-Males','Literate-Males',
'Below-Primary-Males','Below-Middle-Males','Below-Matric-Males','Below-Graduate-Males','Graduate-And-Above-Males'],
 var_name = 'Education-Level', value_name = 'Males')
```

```python
df5 = df2.melt(id_vars=['State-Code','District-Code','Name','TRU'],value_vars=['Total-Females','Illiterate-Females','Literate-Females',
'Below-Primary-Females','Below-Middle-Females','Below-Matric-Females','Below-Graduate-Females','Graduate-And-Above-Females'],
 var_name = 'Education-Level', value_name = 'Females')
```

```python
df6 = df3[['State-Code','District-Code','Name','TRU','Education-Level','Tot-P']]
df6['Tot-M'] = df4['Males']
df6['Tot-F'] =df5['Females']
df6['Education-Level'] = df6['Education-Level'].str.replace("-Persons","")
df6['Education-Level'] = df6['Education-Level'].str.replace("Below-Primary","Literate but below primary")
df6['Education-Level'] = df6['Education-Level'].str.replace("Below-Middle","Primary but below middle")
df6['Education-Level'] = df6['Education-Level'].str.replace("Below-Matric","Middle but below matric/secondary")
df6['Education-Level'] = df6['Education-Level'].str.replace("Below-Graduate","Matric/Secondary but below graduate")
df6['Education-Level'] = df6['Education-Level'].str.replace("Graduate-And-Above","Graduate and above")
df6.drop('District-Code',axis=1,inplace=True)
```

```python
df6
```

```python
df7 = pd.merge(df1, df6, on=['Name','TRU','Education-Level']).drop(columns=['State-Code_x']).rename(columns={'State-Code_y':'State-Code'})
df7
```

```python
df7 = df7.apply(pd.to_numeric, errors='ignore')
df7.dtypes
```

```python
df7['Persons-exactly-1'] = df7['Tot-P']-df7['Persons-2']
df7['Males-exactly-1'] = df7['Tot-M']-df7['Males-2']
df7['Females-exactly-1'] = df7['Tot-F']-df7['Females-2']
df7
```

```python
df7['Persons-exactly-2'] = df7['Persons-2']-df7['Persons-3']
df7['Males-exactly-2'] = df7['Males-2']-df7['Males-3']
df7['Females-exactly-2'] = df7['Females-2']-df7['Females-3']
df7
```

```python
df7['Persons-atleast-3'] = df7['Persons-3']
df7['Males-atleast-3'] = df7['Males-3']
df7['Females-atleast-3'] = df7['Females-3']
df7
```

```python
df7.sort_values(['State-Code','TRU'])
```

# Final Dataframe

```python
final_df = df7[['State-Code','Name','TRU','Education-Level','Tot-P','Tot-M','Tot-F','Persons-exactly-1','Males-exactly-1','Females-exactly-1','Persons-exactly-2','Males-exactly-2','Females-exactly-2','Persons-atleast-3','Males-atleast-3','Females-atleast-3']]
final_df.sort_values(['State-Code','Name','TRU'],inplace=True)
final_df
```

```python
final_df.to_csv("cleaned_datafile2.csv",index=False)
```
